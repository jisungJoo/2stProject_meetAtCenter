### (같이하기)작성자, 댓글 가져오기 함수로 만들기

```python
def comment_def():
    #현재 페이지로 긁어오기
    soup = bs(driver.page_source,'html.parser')
    
    #작성자
    writer = soup.find('a',class_='FPmhX notranslate nJAzx')

    #댓글작성자, 댓글내용
    commenters = soup.findAll('a',class_="FPmhX notranslate TlrDj")
    comments = soup.select('._6lAjh + span')

    for index in range(len(commenters)):
        print('작성자 :',commenters[index].text)
        print('댓글 :',comments[index].text)
        print('-'*30)
```

# 해시태그만 긁어오기

```python
def comment_hash_def():
    #현재 페이지로 긁어오기
    soup = bs(driver.page_source,'html.parser')
    
    #모든 댓글 긁어오기 해시태그 위치가 항상 다름
    comments = soup.select('._6lAjh + span')
    
    for comment in comments:
        #띄어쓰기 기준으로 자르기
        words = comment.text.split()
        
        for word in words:
            #'#'이 있는것만 가져오기
            if '#' in word:
                print(word)
```

```python
comment_hash_def()
```

```python
for i in range(5):
    nextMove()
    comment_hash_def()
```

```python
url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EA%BC%AC%EB%A0%9B'

driver = wb.Chrome()
driver.get(url)
```

```python
body = driver.find_element_by_tag_name('body')

for num in range(50):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
```

```python
soup = bs(driver.page_source, 'lxml')
```

```python
img_src = soup.select('img._img')
time.sleep(2)
```

```python
len(img_src)
```

```python
img_src2 = []

for i in img_src :
    img_src1 = i['src']
    img_src2.append(img_src1)
```

```python
img_src2
```

```python
len(img_src2)
```

```python
file_no = 0

for i in range(len(img_src2)):
    
    try:
        urlretrieve(img_src2[i], './img/' + str(file_no) + '.jpg')
    except :
        continue
    
    file_no += 1
    time.sleep(2)
    
    print('%s번째 사진 저장중입니다.'%file_no)
```
